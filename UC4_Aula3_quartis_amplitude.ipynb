# Interpretação de Quartis

---

## O que os quartis nos mostram?

Os quartis ajudam a responder:

> Como os dados estão distribuídos?

Eles dividem os dados em 4 partes:

- Q1 → 25% dos dados estão abaixo desse valor  
- Q2 → 50% dos dados (mediana)  
- Q3 → 75% dos dados  

---



## O que observar na análise

### 1. Concentração dos dados

Observe os intervalos:

- mínimo → Q1  
- Q1 → Q2  
- Q2 → Q3  
- Q3 → máximo  

Regra:

- Intervalo pequeno → dados concentrados  
- Intervalo grande → dados espalhados  

---

### 2. Assimetria dos dados

Compare:

- distância entre Q1 e Q2  
- distância entre Q2 e Q3  

Interpretação:

- Distâncias parecidas → dados equilibrados  
- Um lado maior → dados inclinados (assimétricos)  

Exemplo:

Q1 = 10  
Q2 = 12  
Q3 = 30  

→ dados puxados para valores maiores  

---

### 3. Presença de outliers

Observe:

- distância entre Q3 e o valor máximo  
- distância entre o valor mínimo e Q1  

Se houver grande diferença:

→ possível presença de valores extremos (outliers)

---

### 4. Variabilidade dos dados

Use o intervalo interquartil (IQR):

IQR = Q3 - Q1

Interpretação:

- IQR pequeno → dados homogêneos  
- IQR grande → dados muito variados  



---

## O que podemos concluir com quartis

### Exemplo 1 – Dados concentrados

Q1 = 10  
Q2 = 12  
Q3 = 14  

Conclusões:

- dados próximos  
- baixa variação  
- comportamento consistente  

---

### Exemplo 2 – Dados espalhados

Q1 = 10  
Q2 = 20  
Q3 = 50  

Conclusões:

- alta variabilidade  
- dados muito diferentes entre si  
- comportamento instável  

---

### Exemplo 3 – Presença de outlier

Q1 = 10  
Q2 = 12  
Q3 = 14  
máximo = 100  

Conclusões:

- existe valor extremo  
- média pode estar distorcida  
- mediana é mais confiável  

---

- detectar outliers

- entender a distribuição  
- preparar dados para modelos  
- criar faixas de classificação (baixo, médio, alto)
