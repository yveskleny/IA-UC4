```python
import numpy as np
import statsmodels.formula.api as smf
import pandas as pd
```

```python
df = pd.read_spss("alunos.sav")
df.head()
```

```python
model = smf.ols(formula="raclog ~ grupo", data=df).fit()
model.summary()
```
